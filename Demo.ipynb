```python
import numpy as np

from segments import Segment
```

## Create Segments

```python
segment1 = Segment(1.0)     # First Segment with length 1.0 m (will serve as proximal segment)
segment2 = Segment(0.75)    # Second Segment with length 0.75 m
segment3 = Segment(0.5)     # Third Segment with length 0.5 m
```

## Link Segments

```python
segment1.link_segment(segment2)    # Link segment2 to segment1
segment2.link_segment(segment3)    # Link segment3 to segment2
```

## Set Origin Position of Proximal Segment
Default is (0,0)

```python
segment1.set_prox_pt((0,0))
```

## Set Segment Angles
Global Angles in Degrees

```python
segment1.set_angle(45)
segment2.set_angle(0)
segment3.set_angle(-45)
```

## Plot Wireframe

Plots the wireframe of the segment it was called on and all distal segments

```python
segment1.plot_wireframe()
```

## Iterate of Angle Series to Create Animates

Note: I usually save an image for each time point, then load the images into a gif-making program

```python
# Reset Angles
segment1.set_angle(0)
segment2.set_angle(0)
segment3.set_angle(0)
```

```python
# Set Segment Colors
segment2.color = 'b'
segment3.color = 'r'
```

```python
nSamples = 10
seg1_angles = np.linspace(0,90,nSamples)
seg2_angles = np.linspace(-45,90,nSamples)
seg3_angles = np.linspace(-90,90,nSamples)
```

```python
for i in range(nSamples):
    segment1.set_angle(seg1_angles[i])
    segment2.set_angle(seg2_angles[i])
    segment3.set_angle(seg3_angles[i])
    
    #segment1.plot_wireframe(title=f'T = {i}',show_joints=True,axis_range=(-2,2))
    
    # Output Image for Each Frame
    segment1.plot_wireframe(title=f'T = {i}',show_joints=True,axis_range=(-2,3),legend=['Seg 1','Seg 2','Seg 3'],
                            show=False,output=True,output_path=f'figure_{i}.png')
```
